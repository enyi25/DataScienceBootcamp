```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv(r"C:\Dev\GIT\DataScience_Bootcamp\Week3\Earthquake_Dataset\all_month.csv")
```

```python
# 1. Bar Graph: Count of earthquakes by magnitude range
data['mag_range'] = pd.cut(data['mag'], bins=[0, 2, 4, 6, 8, 10], labels=['0-2', '2-4', '4-6', '6-8', '8-10'])
mag_counts = data['mag_range'].value_counts().sort_index()
```

```python
plt.figure(figsize=(8, 6))
mag_counts.plot(kind='bar', edgecolor='black')
plt.title('Count of Earthquakes by Magnitude Range')
plt.xlabel('Magnitude Range')
plt.ylabel('Number of Earthquakes')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 2. Pie Chart: Distribution of earthquake types
type_counts = data['type'].value_counts()

plt.figure(figsize=(8, 8))
type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, legend=True, explode=[0.1] + [0] * (len(type_counts) - 1))
plt.title('Distribution of Earthquake Types')
plt.ylabel('')  # Hide the y-label
plt.show()
```

```python
# 3. Box Plot: Depth distribution by earthquake type
plt.figure(figsize=(10, 6))
data.boxplot(column='depth', by='type', grid=False, showfliers=False)
plt.title('Depth Distribution by Earthquake Type')
plt.suptitle('')  # Remove default matplotlib title
plt.xlabel('Earthquake Type')
plt.ylabel('Depth (km)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Line Chart: Number of earthquakes over time
data['time'] = pd.to_datetime(data['time'])
earthquake_counts = data.set_index('time').resample('D').size()
```

```python
plt.figure(figsize=(12, 6))
earthquake_counts.plot(kind='line', marker='o', linewidth=0.75)
plt.title('Number of Earthquakes Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Earthquakes')
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 5. Scatter Plot: Magnitude vs. Depth
plt.figure(figsize=(8, 6))
plt.scatter(data['mag'], data['depth'], alpha=0.5)
plt.title('Magnitude vs. Depth of Earthquakes')
plt.xlabel('Magnitude')
plt.ylabel('Depth (km)')
plt.grid(alpha=0.7)
plt.show()
```

```python
# 6. Geographical Scatter Plot: Latitude vs Longitude
plt.figure(figsize=(10, 8))
plt.scatter(data['longitude'], data['latitude'], c=data['mag'], cmap='viridis', alpha=0.7)
plt.colorbar(label='Magnitude')
plt.title('Geographical Distribution of Earthquakes')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(alpha=0.5)
plt.show()
```

```python
# Extract high-magnitude earthquake data
high_magnitude = data[data['mag'] > 5][['latitude', 'longitude', 'place', 'mag']]
high_magnitude_sorted = high_magnitude.sort_values(by='mag', ascending=False)
```
